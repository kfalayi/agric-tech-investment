```python
## How much investment per solution focus?
df.groupby(['Solution Focus']).Funding.sum().to_frame().reset_index()
```

```python
## Percentage of all investment in agritech going to biotech companies
df[df['Solution Focus'] == 'Biotech'].Funding.sum() / df.Funding.sum() * 100
```

```python
#How much investment in Biotech (the area which is attracting investors the most)per continent?
df[df['Solution Focus'] == 'Biotech'].groupby('region').Funding.sum().to_frame().reset_index()
```

```python
#Whhich areas are attracting the most investment in Africa?

Africa.groupby(['Solution Focus']).Funding.sum().to_frame().reset_index()
```
